```python
import pandas as pd
```

```python
df = pd.DataFrame()
for i in range(1,7):
    url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2022-0{i}"+".parquet"
    tmp = pd.read_parquet(url,engine='pyarrow')
    df = pd.concat([df,tmp])
del tmp
```

```python
filepath = "./data/yellow_tripdata_2022_jan2jun.parquet"
df.to_parquet(filepath)
```
